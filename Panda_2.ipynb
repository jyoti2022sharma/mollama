#Concat Dataframes with axis argument

This we do when we want to join two dataframe

```python
india_weather = pd.DataFrame({
    "city": ["mumbai","delhi","banglore"],
    "temperature": [32,45,30],
    "humidity": [80, 60, 78]
})
print(india_weather)

us_weather = pd.DataFrame({
    "city": ["new york","chicago","orlando"],
    "temperature": [21,14,35],
    "humidity": [68, 65, 75]
})
print(us_weather)
```

```python
df_con = pd.concat([india_weather,us_weather])
df_con # here index is not in continuation, it is like 0,1,2,0,1,2
```

do continuing the index we need to pass a function ignore index = True

```python
df_con = pd.concat([india_weather,us_weather], ignore_index=True, axis =0)
df_con # here index is not in continuation, it is like 0,1,2,0,1,2
```

```python
df_con = pd.concat([india_weather,us_weather], ignore_index=True, axis =1)
df_con # here index is not in continuation, it is like 0,1,2,0,1,2
```

```python
df_con = pd.concat([india_weather,us_weather], keys = ["india","us"])
df_con
```
